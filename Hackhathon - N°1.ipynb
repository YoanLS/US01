```python
# -*- coding: utf-8 -*-

# Ici, on va extraire l'ensemble des restaurants italiens et français de la base de donnée dianping pour les tracer ensuite sur carto

import sqlite3

# Connect to the database
conn = sqlite3.connect('D:\\UTC\\Info\\US01\\Hackathon\\UTSEUS-shanghai-dianping.db')
c1 = conn.cursor()
c2 = conn.cursor()


# SQL request         
c1.execute('SELECT name, address, latitude, longitude, avg_rating, avg_price, category '+
          'FROM venues, venue_categories '+
          'WHERE venues.business_id = venue_categories.business_id '+
          'AND venue_categories.category = "法国菜" ')

venues1 = c1.fetchall()

c2.execute('SELECT name, address, latitude, longitude, avg_rating, avg_price, category '+
          'FROM venues, venue_categories '+
          'WHERE venues.business_id = venue_categories.business_id '+
          'AND venue_categories.category = "意大利菜" ')

venues2 = c2.fetchall()


'''# Extract restaurants from Baoshan district
df_baoshan = df[['宝山区' in adr for adr in df['address']]]

 Filter restaurants with 0 score
df_baoshan = df_baoshan[(df_baoshan['product_score'] > 0) &
                        (df_baoshan['decoration_score'] > 0) & 
                        (df_baoshan['service_score'] > 0)]###

df_baoshan.to_csv('baoshan.csv')'''
```
